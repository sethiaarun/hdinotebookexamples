## AQE Example Dynamic Coalescing Shuffle Partitions
Configure AQE using Configure Magic Command

```scala
%%configure -f
{ "conf": {"spark.sql.adaptive.enabled": "true",
           "spark.sql.adaptive.coalescePartitions.enabled":"true"
          }
}
```

```scala
case class PageHit(clickId:Int, timeSpent:Int, page:String)
val pageHitData = List(PageHit(1,40,"home"),PageHit(2,30,"checkout"),PageHit(3,25,"home"),
                      PageHit(4,130,"search"),PageHit(5,110,"catalog"),PageHit(6,85,"home"))
val pageHitDF = spark.sqlContext.createDataFrame(pageHitData)
val pageHitSummary = pageHitDF.groupBy("page").agg(sum("timeSpent")).alias("duration")
pageHitSummary.explain
pageHitSummary.foreach(row=>{
    val pageName = row.getString(0)
    val ts = row.getLong(1)
    println(s"${pageName}--${ts}")
})
pageHitSummary.take
```
